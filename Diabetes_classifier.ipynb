```python
# Create a machine learning model that can classify diabetes data set
```

# fetch the data

```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

# analyse the data

```python
df.Outcome.value_counts()
```

# identify the features and the target

```python
target = df.Outcome
features = df.drop(columns=['Outcome'])
```

```python
features.head()
```

# choose the ML algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

# train the model

```python
model.fit(features,target)
```

# find the score

```python
model.score(features,target)
```

# use the model to classify

```python
model.predict([[0,137,40,35,168,43.1,2.288,33]])
```

```python
model.predict(features)
```

```python
df['Predicted_Outcome'] = model.predict(features)
```

```python
df.head()
```

```python
df['Error']=df.Predicted_Outcome == df.Outcome
```

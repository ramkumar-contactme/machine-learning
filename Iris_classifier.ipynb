```python
# Create a machine learning model that can classify iris data set
```

# fetch the data

```python
import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

# analyse the data

```python
df.Species.value_counts()
```

# identify the features and the target

```python
target = df.Species
features = df.drop(columns=['Species','Id'])
```

```python
features.head()
```

# choose the ML algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

# train the model

```python
model.fit(features,target)
```

# find the score

```python
model.score(features,target)
```

# use the model to classify

```python
model.predict(features)
```
